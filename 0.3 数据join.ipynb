```python
def extract_date(x):
    if isnan(x):
        return x
    if len(x) == 8 and x.isdigit():
        y, m, d = x[:4], x[4:6], x[6:]
        return f'{int(y)}-{int(m):02d}-{int(d):02d}'
    if x.replace('/', '').isdigit():
        y, m, d = x.split('/')
        return f'{int(y)}-{int(m):02d}-{int(d):02d}'
    if x.replace('-', '').isdigit():
        y, m, d = x.split('-')
        if len(d) == 4 and len(y) == 2:
            y, d = d, y
        return f'{int(y)}-{int(m):02d}-{int(d):02d}'
```

# 结果过滤

```python
valid_df = join_df.copy()
def eval_to_float(x):
    x = x.strip()
    try:
        return eval(x)
    except:
        return float('nan')
    
def percent_to_float(x):
    x = x.strip()
    try:
        return float(x) / 100
    except:
        return float('nan')
    
valid_df['floss'] = valid_df['floss'].map(lambda x: x.replace('O', '0').replace('x', 'X').replace('X', ''))
valid_df['floss'] = valid_df['floss'].map(eval_to_float)

valid_df['FP'] = valid_df['FP'].map(lambda x: x.replace(':', '').replace('x', 'X').replace('X', '').replace('%', ''))
valid_df['FP'] = valid_df['FP'].map(percent_to_float)

valid_df['FN'] = valid_df['FN'].map(lambda x: x.replace(':', '').replace('关', '').replace('x', 'X').replace('X', '').replace('%', ''))
valid_df['FN'] = valid_df['FN'].map(percent_to_float)

valid_df = valid_df.dropna(subset=['floss', 'FP', 'FN'])
valid_df = valid_df[(valid_df['floss'] < 0.2) & (valid_df['FP'] < 0.15) & (valid_df['FN'] < 0.33)]
print_df(valid_df)
```
