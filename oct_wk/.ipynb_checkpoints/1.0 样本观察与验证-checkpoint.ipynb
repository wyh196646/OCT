```python
class ToMultilabel(DisplayedTransform):
    def __init__(self, label_info):
        self.label_info = label_info
        store_attr()

    def setups(self, dsets):
        return

    def encodes(self, o):
        try:
            return TensorCategory(self.vocab.o2i[o])
        except KeyError as e:
            raise KeyError(f"Label '{o}' was not included in the training dataset") from e
    def decodes(self, o): return Category      (self.vocab    [o])

def MultilabelBlock(label_info):
    return TransformBlock(type_tfms=ToMultilabel(label_info=label_info))
```

```python
root_path = '/home/octusr2/projects/data_fast/proceeded/cp_projection/512'
label_cols = list(filter(lambda x: x.startswith('label_'), df.columns))
df['valid']
dblock = DataBlock(blocks=(ImageBlock, MultilabelBlock),
                           get_x=ColReader('image_path', pref=pref, suff=suff),
                           get_y=ColReader(label_col, label_delim=label_delim),
                           splitter=splitter,
                           item_tfms=item_tfms,
                           batch_tfms=batch_tfms)
dls = ImageDataLoaders.from_dblock(dblock, df, path=path)
```
